```python
cat_counts=data['Genre'].value_counts()
```

```python
cat_counts
```

# I can find top 5 genre from this

# now create bar chart

```python
x_bar=cat_counts.index
```

```python
x_bar
```

```python
y_bar=cat_counts.values
```

```python
y_bar
```

```python
plt.bar(x_bar,y_bar)
```

```python
plt.figure(figsize=(12,8))
plt.bar(x_bar,y_bar)
```

```python
plt.figure(figsize=(12,8))
plt.bar(x_bar,y_bar,width=0.2,color='orange')
plt.xticks(rotation=90,fontsize=10)
plt.xlabel('Genre',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.title('Top Genre Analysis',fontsize=18)
plt.show()
```

```python
sns.countplot(x='Genre',data=data,order=data['Genre'].value_counts().index,color='green')
```

# create pie chart

```python
sales_data=data[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']]
```

```python
sales_data
```

```python
region_sales=sales_data.T.sum(axis='columns')
```

```python
region_sales
```

```python
plt.pie(region_sales,labels=region_sales.index,startangle=90,explode=(0.2,0,0,0),autopct='%1.1f%%')
plt.title('Total sales across various regions')
plt.show()
```
